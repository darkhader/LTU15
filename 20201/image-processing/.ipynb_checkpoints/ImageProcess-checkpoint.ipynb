```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
def printInfo(img):
    print('Image shape:\t',img.shape)
    print('Image size:\t',img.size)
    print('Image data type:\t',img.dtype)
    print('======')
 
```

```python
img = cv2.imread('data/hieu.jpg')
printInfo(img)
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img,cv.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img,cv.COLOR_BGR2HSV)
img_gray = cv2.cvtColor(img,cv.COLOR_BGR2GRAY)


print('brg\t',img[400,400])
print('rgb\t',img_rgb[400,400])
print('hsv\t',img_hsv[400,400])
print('hsv\t',img_gray[400,400])

figure(num=None, figsize=(10, 10), dpi=100)

plt.subplot(141)
plt.imshow(img)
plt.title('brg')

plt.subplot(142)
plt.imshow(img_rgb)
plt.title('rgb')

plt.subplot(143)
plt.imshow(img_hsv)
plt.title('hsv')

plt.subplot(144)
plt.imshow(img_gray,cmap='Greys_r')
plt.title('gray')

print('bgr img info')
printInfo(img)
print('rgb img info')
printInfo(img_rgb)
print('hsv img info')
printInfo(img_hsv)
print('gray img info')
printInfo(img_gray)

plt.show()
```

```python
def BGR2CMYK(img):
    cmyk = []
    for line in img:
        new_line = []
        for pixel in line:
            k = min(pixel)
            c = 255 - pixel[2] - k
            m = 255 - pixel[1] - k
            y = 255 - pixel[0] - k
            new_line.append([c,m,y,k])
        cmyk.append(new_line)
    return np.asarray(cmyk)
def BGR2CMY(img):
    cmy = []
    for line in img:
        new_line = []
        for pixel in line:
            c = 255 - pixel[2]
            m = 255 - pixel[1]
            y = 255 - pixel[0]
            new_line.append([c,m,y])
        cmy.append(new_line)
    return np.asarray(cmy)
def CMYK2RGB(cmyk):
    img = []
    return img
def CMY2RGB(cmy):
    img = []
    return img
```
