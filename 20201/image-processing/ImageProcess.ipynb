```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
def printInfo(img):
    print('Image shape:\t',img.shape)
    print('Image size:\t',img.size)
    print('Image data type:\t',img.dtype)
    print('======')
```

```python
img = cv2.imread('data/hieu.jpg')
printInfo(img)
plt.imshow(img)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)


print('brg\t',img[400,400])
print('rgb\t',img_rgb[400,400])
print('hsv\t',img_hsv[400,400])
print('hsv\t',img_gray[400,400])

figure

(num=None, figsize=(10, 10), dpi=100)

plt.subplot(141)
plt.imshow(img)
plt.title('brg')

plt.subplot(142)
plt.imshow(img_rgb)
plt.title('rgb')

plt.subplot(143)
plt.imshow(img_hsv)
plt.title('hsv')

plt.subplot(144)
plt.imshow(img_gray,cmap='Greys_r')
plt.title('gay')

print('bgr img info')
printInfo(img)
print('rgb img info')
printInfo(img_rgb)
print('hsv img info')
printInfo(img_hsv)
print('gray img info')
printInfo(img_gray)

plt.show()
```
